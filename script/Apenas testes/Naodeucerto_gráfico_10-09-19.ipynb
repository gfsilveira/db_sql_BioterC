```python
#Filtrar apenas as linhagens de interesse que são utilizadas atualmente no biotério
ex_ling = [
    'Test',
    'AAA',
    'Swiss',
    'C57BL06WT\t',
    'Swiss/CFW B',
    'Swiss/CFW D',
    'Swiss/CFW A',
    'Swiss/CFW C',
    'Swiss/CFW',
    'Swiss-2',
    'Swiss-2 A ',
    'NZB/BINJ',
    'NZW/LacJ',
    'NOD / SHILTJ'
]
    
for l in ex_ling:
    tirar_ling = tab_est_est.query("ling != ' + l + '")
tab_est_est.head()
```

```python
#Filtrando apenas os animais que realmente nasceram e numero de nascidos inferior a 200

tirar_ling = tab_est_est[(tab_est_est['util'] == 'nasc') & (tab_est_est['num_nasc'] < 200)]
tirar_ling.head()
```

```python
#alterando as datas de segundos para datas
datas = ['date_hoje', 
         'date_nasc', 
         'data_inicio', 
         'date_acas', 
         'date_nasc_acas', 
         'date_sex', 
         'data_solic', 
         'data_forn', 
         'data_fim']

for d in datas:
    tirar_ling[d] = pd.to_datetime(tirar_ling[d], unit = 's') #indica o formato das datas
    
tirar_ling.head()
```
